```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import style
style.use('ggplot')
plt.grid(True,color='k')
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#DIST_PLOT
```

```python
sns.distplot(tips['tip'])
```

```python
sns.distplot(tips['total_bill'], kde = False,color = 'k' )
```

```python
sns.distplot(tips['total_bill'], kde = False, bins =50, color ='red')
```

```python
#jointplot - helps in comparing two variables in a single plot
```

```python
sns.jointplot(x='total_bill',y= 'tip',data = tips).set_axis_labels("x",'y')
plt.grid(True,color = 'k')
```

```python
#we can use kind attribute for different plot by default it is scatter
#such as kind =“scatter” | “reg” | “resid” | “kde” | “hex”
```

```python
sns.jointplot(x= 'total_bill',y='tip', data=tips,kind ='hex')
 
```

```python
sns.jointplot(x= 'total_bill',y='tip', data=tips,kind ='reg')
```

```python
sns.jointplot(x= 'total_bill',y='tip', data=tips,kind ='kde')
```

```python
#we can also normalise the scale using STANDARD_SCALE
```

```python
#GRIDS
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris['species'].unique ()
```

```python
#PairGrid() which is same as pairplot but provides more control over plots
#unlike pairplot which usually provides scatter and histogram
```

```python
g = sns.PairGrid(iris)
```

```python
# we can custom pairgrid like diagonal, upper half and lower half
```

```python
# first we can add type of plot in whole pairgrid
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter,color='k')
```

```python
g= sns.PairGrid(data=iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
#facetGrid
```

```python
#FacetGrid is used to plot grids based on row and column wise values 
#we can use plots to visualise the separated values
```

```python
T = sns.FacetGrid(data=tips, col='time',row='smoker')
T.map(sns.distplot,'total_bill')
```

```python
#we can also use two parameter plots such as scatter plot
```

```python
T = sns.FacetGrid(data=tips, col='time',row='smoker')
T.map(plt.scatter, 'total_bill', 'tip', marker='v')
```
